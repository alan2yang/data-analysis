```python
import pandas as pd
from pymongo import MongoClient
import numpy as np
import matplotlib.pyplot as plt

% matplotlib inline

plt.rcParams['figure.figsize'] = (15, 10)
```

```python
client=MongoClient()
collection=client['51jb']['szjobs']

data=pd.DataFrame(list(collection.find()))
data=data[['company','job','location','publish_time','salary','require']]
data
```

```python
data_ana=data[data['job'].str.contains('数据分析')]
data_ana
```

```python
# 如何批量处理列表？？

# type(data_ana['salary'][0][0])
# data_ana['require'][0][0]

data_ana['l_price']=data_ana['salary'].str[0]
data_ana['h_price']=data_ana['salary'].str[1]
data_ana['m_price']=(data_ana['l_price']+data_ana['h_price'])/2
# data_ana['l_price'].astype(np.float16)
# data_ana['h_price'].astype(np.float16)

data_ana['work_year']=data_ana['require'].str[0]
data_ana['education']=data_ana['require'].str[1]
data_ana['need_person']=data_ana['require'].str[2]


print(data_ana.dtypes)

c_data=data_ana.dropna()

c_data=c_data[~c_data['education'].str.contains('招')]

c_data
```

```python
# 工作年限与工资的关系

g_work_year=c_data.groupby('work_year')


# # 工作年限的平均薪资
# salary_by_work_year=g_work_year['m_price'].mean()
# # 工作年限的岗位数量
# num_by_work_year=g_work_year['m_price'].count()


# 不同工作年限下，薪资的分布情况
sta_by_work_year=g_work_year['m_price'].describe()
print(sta_by_work_year)

# 不同工作年限间的薪资对比
ax=plt.subplot(111)
ax2=ax.twinx()
sta_by_work_year[['mean','min','max']].plot(kind='bar',ax=ax)
# sta_by_education[['count']].plot(secondary_y=True,style='--ko',ax=axes,ylim=(0,160))
sta_by_work_year[['count']].plot(style='--ko',ax=ax2,ylim=(0,160))
ax.legend(loc=2)
ax2.legend(loc=1)


group_nums=len(g_work_year)

fig,axes=plt.subplots(group_nums,1,figsize=(10,40))
i=0
for g_name,g_content in g_work_year:
    g_content['m_price'].hist(bins=30,histtype='bar',ax=axes[i],edgecolor='k')
    axes[i].set_title(g_name)
    axes[i].set_ylim([0,12])
    axes[i].set_xlabel('薪资')
    axes[i].set_ylabel('公司数')

    i+=1
```

```python
# 薪资与学历的关系

g_education=c_data.groupby('education')

sta_by_education=g_education['m_price'].describe()
print(sta_by_education)

# 不同学历间的薪资对比
ax=plt.subplot(111)
ax2=ax.twinx()
sta_by_education[['mean','min','max']].plot(kind='bar',ax=ax)
# sta_by_education[['count']].plot(secondary_y=True,style='--ko',ax=axes,ylim=(0,160))
sta_by_education[['count']].plot(style='--ko',ax=ax2,ylim=(0,160))
ax.legend(loc=2)
ax2.legend(loc=1)

# 不同学历下薪资和公司数量的关系
group_nums=len(g_education)
fig,axes=plt.subplots(group_nums,1,figsize=(10,40))
i=0
for g_name,g_content in g_education:
    g_content['m_price'].hist(bins=30,histtype='bar',ax=axes[i],edgecolor='k')
    axes[i].set_title(g_name)
    axes[i].set_ylim([0,30])
    axes[i].set_xlabel('薪资')
    axes[i].set_ylabel('公司数')
    i+=1
```
